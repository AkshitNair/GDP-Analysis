Removing data for duration '2016-17' from dataframe

```python
df2 = DF[DF.Duration != '2016-17']
```

Creating dataframe for Growth over previous year and for duration from 2013-2016 to calculate the mean value

```python
df3 = df2.loc[(df2['Items  Description'] == '(% Growth over previous year)') & (df2['Duration'].isin(['2013-14','2014-15','2015-16']))]
```

Removing the Items  Description column in dataframe and setting Duration as the index

```python
df3=df3.drop(['Items  Description'], axis = 1)
df4=df3.set_index('Duration')
```

Applying Transpose

```python
df6 = df4.transpose()
```

Creating a single data frame for all the states

```python
import os
files = os.listdir('Gross State Value Added (GSVA)/')#importing files from folder and assigning it to variable
```

```python
path = []#creating a list
for i in files:
    path.append(str('Gross State Value Added (GSVA)/'+i))#appending 
```

```python
i = 0#To create a single dataframe
for p in path:
    df = pd.read_csv(p,encoding="ISO-8859-1")#to handle encoding 
    
    if i==0:
        df_gsvp = df.loc[::,['S.No.','Item','2014-15']]
        df_gsvp.columns = ['S.No.','Item',p.split('/')[1].split('-')[1]]
        i = i + 1
    else:
        df_gsvp[p.split('/')[1].split('-')[1]] = df.loc[::,['2014-15']]
```

```python
pd.options.display.max_columns = None
display(df_gsvp)
```

```python
df2_gsvp=df_gsvp.transpose()
df2_gsvp
```

```python
df3_gsvp = df2_gsvp.iloc[1:]# to remove S.No
df3_gsvp
```

To setup columns

```python
df3_gsvp.columns = df3_gsvp.iloc[0]
df3_gsvp
```

```python
df3_gsvp=df3_gsvp.iloc[1:,]
```
